## 2. Preprocessing Pipeline (Reference)

The following steps were applied to produce the data files:

1. **Text extraction**: Raw documents (articles, podcasts, legal filings) were extracted and transcribed (OpenAI Whisper for audio)
2. **Chunking**: Documents segmented into ~250-word chunks with 25-word overlap
3. **Normalization**: Phrase normalization ("AI" → "artificial_intelligence"), stopword removal, lowercasing
4. **Artist probes**: Lovato et al. (2024) survey responses converted to declarative Likert-anchored statements
5. **Likert anchors**: 250 synthetic anchor statements generated via GPT-5.1 (5 themes × 5 levels × 10 styles)

See `data/README.md` for the full data dictionary.
